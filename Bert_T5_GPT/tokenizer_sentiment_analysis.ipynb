```python
test_sentences = ['today is not that bad','today is so bad']
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model_name = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
batch_input = tokenizer(test_sentences,return_tensors='pt', padding=True, truncation=True,max_length=256)
```

```python
tokenizer(test_sentences[0],)
```

```python
ids_encode=tokenizer.encode(test_sentences[0],)
```

```python
tokenizer.tokenize(test_sentences[0],)
ids = tokenizer.convert_tokens_to_ids(tokenizer.tokenize(test_sentences[0],))
decode = tokenizer.decode(ids_encode)

print(test_sentences[0])
print(ids)
print(ids_encode)
print(decode)
```

```python
print(tokenizer.vocab)
print(tokenizer.vocab_size)
print(tokenizer.special_tokens_map)
print(tokenizer.convert_tokens_to_ids([special for special in tokenizer.special_tokens_map.values()]))
```

```python

#padding='max_length' 补0 ；  padding=True 补到最长的句子长度; padding='longest' 补到最长的句子长度
#truncation=True 截断


'''
truncation=False（默认）	不截断，如果输入超过 max_length 可能会报错
truncation=True	自动截断到 max_length
truncation='longest_first'	从最长的部分开始截断（默认方式）
truncation='only_first'	只截断第一个句子（适用于句子对任务）
truncation='only_second'	只截断第二个句子（适用于句子对任务）
'''
tokenizer(test_sentences,return_tensors='pt', padding=True, truncation=True,max_length=256)
```
